####  REAL TIME TWITTER SENTIMENT PLOT

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook
```

```python
frame_len = 10000 #latest 10k points
fig = plt.figure(figsize = (9,6))

def anime(i):
    data= pd.read_csv("sentiment.csv")
    y1 = data["usa"]
    y2 = data["china"]
    
    if len(y1) <= frame_len:
        plt.cla() #clear axis
        plt.plot(y1,label="usa")
        plt.plot(y2,label="china")
    else:
        plt.cla() #clear axis
        plt.plot(y1[-frame_len],label="usa")
        plt.plot(y2[-frame_len],label="china")
                 
    plt.legend(loc="upper-left")
    plt.tight_layout()
                 

ani = FuncAnimation(plt.gcf(),anime,interval=1000)    #1000ms
```
